```python
import pandas as pd
import re
```

```python
# CSV File

df = pd.read_csv("sales_data_sample.csv",encoding="cp1252")
print(df.head())
```

```python
# clean PRODUCTLINE COLUMN

df["PRODUCTLINE"]=df["PRODUCTLINE"].str.strip().str.title()

df["PRODUCTLINE"].head()
```

```python
#Clean sales column

df["SALES"].dtype
```

```python
df["SALES"]=df["SALES"].astype(str)
df["SALES"]=df["SALES"].apply(lambda x: re.sub(r"[^0-9.]", "",x))

df["SALES"]=df["SALES"].astype(float)
df["SALES"].head()
```

```python
# Dictionary of Product Sales
print("\n==TOTAL SALES BY PRODUCT LINE==\n")
product_sales={}

for index, row in df.iterrows():
    
    product=row["PRODUCTLINE"]
    sales=row["SALES"]
    if product not in product_sales:
        product_sales[product]=sales
    else:
        product_sales[product]+=sales

product_sales
        
```

```python
#TOTAL SALES BY PRODUCTLINE

print("Total Sales By Product Line\n")

for product,total in product_sales.items():
    print(product,":",round(total,2))
```

```python
#Identify high revenue products

print("\nHigh Revenue Products\n")

for product, total in product_sales.items():
    if total>100000:
        print(product,"=> High Revenue Product")
```

```python
#Summary Report

print("\n===== SALES REPORT =====\n")

print("Number of Product Lines:",len(product_sales))
for product,total in product_sales.items():
    print(f"{product}: {total:.2f}")

print("===========================")
```
