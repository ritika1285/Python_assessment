```python
import pandas as pd
```

```python
# DATA 

df = pd.read_csv(
    "sales_data_sample.csv",
    encoding="latin1"
)

print("Dataset Loaded Successfully")
```

```python
# DATA TYPE VALIDATION

numeric_cols = [
    'ORDERNUMBER',
    'QUANTITYORDERED',
    'PRICEEACH',
    'SALES'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(
        df[col],
        errors='coerce'
    )

df['ORDERDATE'] = pd.to_datetime(
    df['ORDERDATE'],
    errors='coerce'
)
```

```python
# MISSING VALUE

missing_before = df.isnull().sum().sum()

df.dropna(
    subset=['SALES','PRICEEACH'],
    inplace=True
)

missing_after = df.isnull().sum().sum()
```

```python
# DUPICATE VALUE

duplicates = df.duplicated().sum()

df.drop_duplicates(inplace=True)
```

```python
# REVENUE VALIDATION

df['ExpectedRevenue'] = (
    df['QUANTITYORDERED']
    * df['PRICEEACH']
)

df['RevenueDifference'] = (
    df['SALES']
    - df['ExpectedRevenue']
)

df['RevenueFlag'] = (
    abs(df['RevenueDifference']) > 1
)
```

```python
# OUTLIER DETECTION

Q1 = df['SALES'].quantile(0.25)
Q3 = df['SALES'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['OutlierFlag'] = (
    (df['SALES'] < lower_bound)
    | (df['SALES'] > upper_bound)
)

outlier_count = df['OutlierFlag'].sum()
```

```python
# CLEAN DATA

df.to_csv(
    "cleaned_sales_data.csv",
    index=False
)

```

```python
print("\n==============================")
print(" SALES INTEGRITY REPORT ")
print("==============================")

print(f"Total Records: {len(df)}")
print(f"Duplicates Removed: {duplicates}")
print(f"Missing Values Before: {missing_before}")
print(f"Missing Values After: {missing_after}")

print(
    f"Revenue Mismatch Records: "
    f"{df['RevenueFlag'].sum()}"
)

print(
    f"Revenue Outliers: "
    f"{outlier_count}"
)

print(
    f"Total Revenue: "
    f"${df['SALES'].sum():,.2f}"
)

print(
    f"Average Revenue: "
    f"${df['SALES'].mean():,.2f}"
)

print(
    "\nClean dataset exported as "
    "'cleaned_sales_data.csv'"
)
```
